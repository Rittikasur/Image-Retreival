```python
#Importing all modules 
import numpy as np       #To numerical calculation
import pandas as pd      #To read csv
import os                #To create folders and get item of folder
import shutil            #To move file from one directory to another 
```

```python
# 1. Read the dataset 
# 2. Drop Unnecesssary Columns 
df= pd.read_csv('data-new-items-July2022-Sept2022.xlsx - New_Image_List.csv')
df.drop(['PERIOD', 'STORECODE', 'ITEMCODE','CATEGORY', 'MANUFACTURE', 'BRAND', 'ITEMDESC', 'MRP', 'PACKSIZE', 'PACKTYPE',
       'COMMENTS','NEW ITEMCODE.1', 'ACTION', 'OPS COMMENT','NEW ITEMCODE' ], axis= 1,  inplace= True)
```

```python
# breaking IMAGE name and remove few parts and saving important part to a list
new = df["IMAGE"].str.split(".", n = 1, expand = True)
new1= new[0].str.split("#", expand = True)
file_names = new1[0].tolist()
```

```python
# Name of all Image file
all_file_names_unprocessed = os.listdir("all_data")
```

```python
#Getting processed name i.e except numbers and extension 

def get_just_names(name):
    return(name[:len(name)-6])   
all_file_names_processed = list(map(get_just_names,all_file_names_unprocessed))
```

```python
def get_file_name_from_index(name):
    return(name[14:])
file_names = list(map(get_file_name_from_index,file_names))
```

```python
file_names
```

```python
# an empty list for unknown file
blacklist = []
```

```python
raw_root = "all_data"   #Source
root = "processed_data"    #Destination

for index,element in enumerate(file_names):
    element = element.strip()
    if(element in blacklist):
        continue
    rslt = [i for i in range(len(all_file_names_unprocessed)) if all_file_names_unprocessed[i].__contains__(element)]
    folder_submane = os.path.join(root,element)
    if(os.path.exists(folder_submane)):
        print(len(os.listdir(folder_submane))>0)
        print(len(rslt) == 0)
    else:
        os.mkdir(folder_submane)
#         print(folder_submane)
#         print("Path CReated ",folder_submane)
        for index_file in rslt:
            tmp_subfolder = all_file_names_unprocessed[index_file]
            from_folder = os.path.join(raw_root,tmp_subfolder)
            to_folder = os.path.join(folder_submane,tmp_subfolder)
#             print(from_folder,to_folder)
            try:
                shutil.move(from_folder,to_folder)
            except Exception as fnf:
                print(fnf)
                print(to_folder)
                print(from_folder)
                print("----------------------------------------------------------------------------")
                blacklist.append(element)
    all_file_names_unprocessed = os.listdir("all_data")
```
